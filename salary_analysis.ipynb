```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
salary_path = "/Users/brewerka/mvp_project/sred_mes.xlsx"
inflation_path = "/Users/brewerka/mvp_project/infl.csv"
```

```python
df_2000_2016 = pd.read_excel(salary_path, sheet_name="2000-2016 гг.", header=2)
```

```python
df_2000_2016
```

```python
df_2017_2023 = pd.read_excel(salary_path, sheet_name="с 2017 г.", header=4)
```

```python
df_2000_2016.rename(columns={df_2000_2016.columns[0]: "Отрасль"}, inplace=True)
df_2017_2023.rename(columns={df_2017_2023.columns[0]: "Отрасль"}, inplace=True)
```

```python
df_2000_2016["Отрасль"] = df_2000_2016["Отрасль"].astype(str).str.replace('\xa0', ' ').str.strip().str.lower()
df_2017_2023["Отрасль"] = df_2017_2023["Отрасль"].astype(str).str.replace('\xa0', ' ').str.strip().str.lower()
```

```python
sectors_1 = {
    "образование": "образование",
    "здравоохранение и предоставление социальных услуг": "здравоохранение",
    "транспорт и связь": "транспорт"
}
```

```python
sectors_2 = {
    "образование": "образование",
    "деятельность в области здравоохранения и социальных услуг": "здравоохранение",
    "транспортировка и хранение": "транспорт"
}
```

```python
df1 = df_2000_2016[df_2000_2016["Отрасль"].isin(sectors_1.keys())].copy()
df1["Сектор"] = df1["Отрасль"].map(sectors_1)
```

```python
df2 = df_2017_2023[df_2017_2023["Отрасль"].isin(sectors_2.keys())].copy()
df2["Сектор"] = df2["Отрасль"].map(sectors_2)
```

```python
df_all = pd.concat([df1, df2], ignore_index=True)
df_all = df_all.drop(columns=["Отрасль"])
```

```python
df_long = df_all.melt(id_vars=["Сектор"], var_name="Год", value_name="Зарплата")
df_long["Год"] = pd.to_numeric(df_long["Год"], errors="coerce")
df_long = df_long.dropna(subset=["Год", "Зарплата"])
df_long["Год"] = df_long["Год"].astype(int)
```

```python
df_cpi = pd.read_csv(inflation_path)
df_cpi.rename(columns={"Всего": "CPI"}, inplace=True)
df_cpi = df_cpi[["Год", "CPI"]]
df_cpi["Год"] = pd.to_numeric(df_cpi["Год"], errors="coerce")
```

```python
df_merged = pd.merge(df_long, df_cpi, on="Год", how="left")
df_merged["Реальная зарплата"] = df_merged["Зарплата"] / (df_merged["CPI"] / 100)
```

```python
df_merged
```

```python
df_merged.to_csv("zarplaty_realnye.csv", index=False, encoding='utf-8-sig')
```

```python
sns.set_theme(style="whitegrid", font_scale=1.1)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_merged, x="Год", y="Зарплата", hue="Сектор", marker="o")
plt.title("Номинальная зарплата по секторам (2000–2024)")
plt.ylabel("Зарплата, ₽")
plt.xlabel("Год")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_merged, x="Год", y="Реальная зарплата", hue="Сектор", marker="o")
plt.title("Реальная зарплата по секторам (с учетом инфляции)")
plt.ylabel("Реальная зарплата, ₽")
plt.xlabel("Год")
plt.tight_layout()
plt.show()
```

```python
df_merged_sorted = df_merged.sort_values(["Сектор", "Год"])
df_merged_sorted["Δ Реальная зарплата"] = df_merged_sorted.groupby("Сектор")["Реальная зарплата"].diff()

plt.figure(figsize=(12, 6))
sns.barplot(data=df_merged_sorted, x="Год", y="Δ Реальная зарплата", hue="Сектор")
plt.axhline(0, color="gray", linestyle="--")
plt.title("Изменение реальной зарплаты по сравнению с предыдущим годом")
plt.ylabel("Изменение, ₽")
plt.xlabel("Год")
plt.tight_layout()
plt.show()
```

Номинальная зарплата по секторам (2000–2024):

* Транспорт стабильно лидирует по уровню зарплат.
* Образование и здравоохранение сильно отстают, но тренд на рост сохраняется.
* Видно резкое ускорение с 2020 года.

Реальная зарплата с учётом инфляции:

* Пик реальной зарплаты — около 2019–2020.
* Далее — резкое падение в 2021–2022, несмотря на рост номинала (виновата инфляция).
* К 2024 году реальная зарплата вернулась к значениям 2015–2016 годов.

Изменение реальной зарплаты год к году:

* Отрицательные пики в 2015, 2020–2021 — снижение покупательной способности.
* Лучший рост — в 2018–2019.
* После 2021 года — резкое замедление или падение.

Выводы:

* Инфляция съедает рост: номинальные зарплаты растут, но реальный рост нестабилен.
* Пандемия и кризисы (2020–2022) резко снизили реальную зарплату.
* Транспорт остаётся наиболее высокооплачиваемым из трёх секторов.
* Образование и здравоохранение системно недофинансируются в реальных значениях.
